### The Scientific Method


1. We always start with a question. Based on your intuition/understandings or bussiness logic.
2. Research- see what people have discovered so far on the topic
    - Prior research by other people
    - Ways of working with our type of data
    - Ideas we can improve on
    - See things that work and don't work so we dont waste our time redoing them.
    - How to translate the task/question from general language to being mathematically modeled
3. Hypothesis - [In statistical hypothesis testing](https://en.wikipedia.org/wiki/Hypothesis), two hypotheses are compared. These are called the null hypothesis and the alternative hypothesis. The null hypothesis is the hypothesis that states that there is no relation between the phenomena whose relation is under investigation, or at least not of the form given by the alternative hypothesis. The alternative hypothesis, as the name suggests, is the alternative to the null hypothesis: it states that there is some kind of relation
4. Test with an Experiment
   - This is the moment we receive the data
   - EDA on the data - validate, transform, explore etc.
   - Ask questions about the data
5. Report your results - weather the hypothesis is True or False we have to report it so other people may know if it it works or not. We don't want them to waste their time recreating our reserach again and again right?
6. Repeat - ask again, test again ...
